```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from scipy.integrate import quad
from scipy.fft import fft, ifft
from scipy.interpolate import interp1d
import scipy.integrate as integrate
from functools import partial
from scipy.optimize import minimize, fsolve, basinhopping
import tqdm
import yfinance as yf
import yahoo_fin.options as ops
import cmath

%matplotlib inline

plt.style.use('ggplot')
sns.set_palette('mako')
sns.set_style('darkgrid')
```

```python
def dclProcess(NumberPaths, N, T, x0, theta, delta):
    Z1 = np.random.normal(0.0, 1.0,[NumberPaths,N])
    X = np.zeros([NumberPaths, N + 1])

    X[:, 0] = x0

    time = np.zeros([N+1])
    dt = T / float(N)
    
    for i in range(0, N):

        X[:,i+1] = -1/theta * X[:,i] * dt + np.sqrt((1 - (X[:, i])**2)/(theta * (delta + 1))) * np.sqrt(dt) * Z1[:,i]
            
        if (X[:,i+1] > 1).any():
            X[np.where(X[:,i+1] > 1)[0], i+1] = 1

        if (X[:,i+1] < -1).any():
            X[np.where(X[:,i+1] < 1)[0], i+1] = -1  
            
        time[i+1] = time[i] + dt

    return time, X
```

```python
time, dclpaths = dclProcess(100, 100, 10, 0.04, 1, 1)
```

```python
plt.subplots(figsize=(10, 5), dpi=100)

for i in range(20):
    plt.plot(time, dclpaths[i], label='DCL')
        
    plt.title('DCL Process', fontsize=16)
    plt.xlabel('time', fontsize=14)
    plt.ylabel('Values', fontsize=14)
#     plt.legend()
plt.show()
```
